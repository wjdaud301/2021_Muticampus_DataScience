````python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from konlpy.tag import Hannanum
from konlpy.tag import Komoran
from konlpy.tag import Okt
from konlpy.tag import Kkma
from konlpy.tag import Mecab
# 한글 폰트 패스로 지정
import matplotlib.font_manager as fm
import re
from collections import Counter
%matplotlib inline
````

````python
spwords = set(STOPWORDS)  # 제외할 단어
# spwords.add('내가')  # 제외하고 싶은 단어 추가
````

````python
# file read
file_content = open("../../Desktop/park1.txt", encoding="utf-8").read() # 

wc2 = WordCloud(max_font_size=200, stopwords=spwords,font_path='C:/Windows/Fonts/HMKBP.TTF',
                background_color='white', width=800, height=800, max_words=100)
wc2.generate(file_content)

plt.figure(figsize=(10, 8))
plt.imshow(wc2)
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()
````

---

````python
nlp = Okt()

nouns = nlp.nouns(file_content)

words = []
for n in nouns:
    if len(n) > 1:
        words.append(n)
count = Counter(words)

most = count.most_common(100)

tags = {}
for n, c in most:
    tags[n] =c

wc =WordCloud(max_font_size=250, stopwords=spwords, font_path='C:/Windows/Fonts/HMKBP.TTF',
                background_color='white', width=311, height=512, max_words=100)

gen = wc.generate_from_frequencies(tags)

plt.figure(figsize=(10, 8))
plt.imshow(gen)
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()
````
