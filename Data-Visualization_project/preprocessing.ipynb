````python
def preprocess(Text):
    Text = Text.replace("소계(도심지역)","도심지역")
    Text = Text.replace('소계(강남지역)','강남지역')
    Text = Text.replace('소계(영등포신촌지역)','영등포신촌지역')
    Text = Text.replace('소계(기타지역)','기타지역')
    return Text

ssi20_21['지역분류'] = preprocess(ssi20_21['지역분류'])
````
