```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.io as pio
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
from plotly.validators.scatter.marker import SymbolValidator
```

```python
# unnamed : 0 column없이 불러오기
ds = pd.read_excel('dataset.xlsx',index_col = 0)
```

```python
ds.info()
```

```python
fig = px.scatter(data_frame=ds, x='분기별', y='공실률')
                #color='지역분류',size='공실률')
                                #hover_data=['sepal_width'])
fig.show()
```

```python
subset_ds = ds[(ds['지역']=='부산')&(ds['년도']==2020)]
```

```python
subset_ds
```

```python
fig = px.bar(subset_ds,
            x='년도',y='공실률')
fig.show()
```

```python
fig = px.scatter(subset_ds,
                x='임대료', y='공실률', size="공실률",
                color='년도', trendline="lowess",# hover_name='country'
                log_x=True, size_max=30)
fig.show()
```

```python
subset_ds1 = ds.query('지역== "서울"')
```

```python
fig = px.histogram(subset_ds1,
                x='분기', y='임대료',
                color='년도',title='2020')#, trendline='lowess')
fig.show()
```
